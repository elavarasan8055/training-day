```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fx(x):
  return 3*x**2 - 3*x + 4

def grad(x):
  return 6*x-3

```

```python
x=np.linspace(-10,10,100)

plt.plot(x,fx(x))
plt.plot(x,grad(x))
plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['y','dy'])
plt.show()
```

```python
lr=0.001
epochs=1000


w_path=np.zeros((epochs,2))
w=np.random.choice(x,1)
print(w)
for i in range(epochs):
  grad_w=grad(w)
  w_path[i][0]=grad_w
  w=w-grad_w*lr
  w_path[i][1]=w


print(w)
fig,p=plt.subplots(1,3,figsize=(12,4))
p[0].plot(x,fx(x))
p[0].plot(x,grad(x))


p[0].plot(w_path[-1,1],grad(w_path[-1,1]),'ro')

for i in range(2):
  p[i+1].plot(w_path[:,i])


```
